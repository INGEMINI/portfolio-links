```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

# Page config
st.set_page_config(page_title="Food Dashboard", layout="wide")

# Load data
df = pd.read_excel("swiggy_data.xlsx")

# Clean data (VERY IMPORTANT 🔥)
df['Price (INR)'] = pd.to_numeric(df['Price (INR)'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Title
st.title("🍔 Food Delivery Analytics Dashboard")

# 🔥 SIDEBAR FILTERS
st.sidebar.header("Filters")

city = st.sidebar.multiselect(
    "Select City",
    df['City'].dropna().unique(),
    default=df['City'].dropna().unique()[:5]
)

category = st.sidebar.multiselect(
    "Select Category",
    df['Category'].dropna().unique(),
    default=df['Category'].dropna().unique()
)

# Apply filters
filtered_df = df[
    (df['City'].isin(city)) &
    (df['Category'].isin(category))
]

# Remove NaN safely
filtered_df = filtered_df.dropna(subset=['Price (INR)', 'Rating'])

# 🔥 KPIs
total_orders = len(filtered_df)
avg_rating = round(filtered_df['Rating'].mean(), 2)
total_sales = int(filtered_df['Price (INR)'].sum())
avg_price = round(filtered_df['Price (INR)'].mean(), 2)
max_price = int(filtered_df['Price (INR)'].max())

# 🔥 KPI LAYOUT
col1, col2, col3, col4, col5 = st.columns(5)

col1.metric("Total Orders", total_orders)
col2.metric("Avg Rating", avg_rating)
col3.metric("Total Sales", total_sales)
col4.metric("Avg Price", avg_price)
col5.metric("Max Price", max_price)

# 🔥 TOP CITIES CHART
top_cities = (
    filtered_df.groupby('City')['Price (INR)']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

fig1 = px.bar(
    top_cities,
    x='City',
    y='Price (INR)',
    title="Top 10 Cities by Sales",
    text_auto=True,
    color='Price (INR)'
)

st.plotly_chart(fig1, use_container_width=True)

# 🔥 CATEGORY PIE CHART
category_counts = filtered_df['Category'].value_counts().reset_index()
category_counts.columns = ['Category', 'Count']

fig2 = px.pie(
    category_counts,
    names='Category',
    values='Count',
    title="Category Distribution"
)

st.plotly_chart(fig2, use_container_width=True)

# 🔥 RATING DISTRIBUTION
fig3 = px.histogram(
    filtered_df,
    x='Rating',
    nbins=20,
    title="Rating Distribution"
)

st.plotly_chart(fig3, use_container_width=True)
```
