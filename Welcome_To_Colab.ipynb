# Sales Data Analysis using SQL, SQLite & Python

## Objective: **bold text**
To analyze sales data and generate insights on customer behavior, regional performance, and product categories.

## Tools Used:
- Python (Pandas)
- SQLite
- SQL
- Looker Studio (Dashboard)

## Workflow:
CSV → SQLite → SQL Queries → Insights → Dashboard

## Key Insights:

- West region has highest number of orders.
- Technology category generates highest profit.
- Standard Class is the most used shipping mode.
- Certain cities contribute significantly to overall sales.

```
import pandas as pd

df = pd.read_csv("csv_data.csv")

df.head()
```

```
df.columns
```

```
import sqlite3

conn = sqlite3.connect("orders.db")

df.to_sql("orders", conn, if_exists="replace", index=False)

print("Database created successfully!")
```

```
query = "SELECT COUNT(*) as total_orders FROM orders"
pd.read_sql_query(query, conn)
```

CORE PROJECT IDEA START

1. Orders by Region



```
query = """
SELECT Region, COUNT(*) as total_orders
FROM orders
GROUP BY Region
ORDER BY total_orders DESC
"""
pd.read_sql_query(query, conn)
```

2. Orders by Category

```
query = """
SELECT Category, COUNT(*) as total
FROM orders
GROUP BY Category
"""
pd.read_sql_query(query, conn)
```

3. Orders by Segment

```
query = """
SELECT Segment, COUNT(*) as total
FROM orders
GROUP BY Segment
"""
pd.read_sql_query(query, conn)
```

4. Ship Mode Distribution

```
query = """
SELECT "Ship Mode", COUNT(*) as total
FROM orders
GROUP BY "Ship Mode"
"""
pd.read_sql_query(query, conn)
```

5. Top Cities

```
query = """
SELECT City, COUNT(*) as total_orders
FROM orders
GROUP BY City
ORDER BY total_orders DESC
LIMIT 5
"""
pd.read_sql_query(query, conn)
```

6. SALES ANALYSIS

```
query = """
SELECT Region, SUM(Sales) as total_sales
FROM orders
GROUP BY Region
ORDER BY total_sales DESC
"""
pd.read_sql_query(query, conn)
```

PROFIT ANALYSIS

```
query = """
SELECT Category, SUM(Profit) as total_profit
FROM orders
GROUP BY Category
ORDER BY total_profit DESC
"""
pd.read_sql_query(query, conn)
```
